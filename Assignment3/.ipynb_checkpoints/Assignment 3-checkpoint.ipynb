```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.spatial.distance import euclidean as euc
```

```python
reference = cv2.imread('reference.png', 1)
test1 = cv2.imread('test.png', 1) #[:,:,::-1]
test2 = cv2.imread('test2.png',1) #[:,:,::-1]
images = [reference, test1, test2]
```

```python
def get_interest_points(img, showAllpoints=True):
    sift = cv2.xfeatures2d.SIFT_create()
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    kp, des = sift.detectAndCompute(gray, None)
    if not showAllpoints:
        kp=kp[:100]
    return cv2.drawKeypoints(gray, kp, reference,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS), kp, des
```

```python
img1, kp1, des = get_interest_points(images[0])
```

```python
def calc_dist_btw_point(p1, p2):
    return euc(p1, p2)
```

```python
def matching(img1, img2):
    
    img1, kp1, des1 = get_interest_points(img1)
    img2, kp2, des2 = get_interest_points(img2)
    match = []


    for j in range(0, len(kp1)):
        
        kp_diff = []
        
        for i in range(0, len(kp2)):
            kp_diff.append(euc(des1[j], des2[i].flatten()))

        np_kp = np.array(kp_diff)

        min_diff = np.min(np_kp)
        second_min = np.min(np.delete(np_kp, [np.argmin(np_kp)]))

        if np.divide(min_diff, second_min) < 0.8:

            des2arg = np.argmin(kp_diff)
            
            match.append((kp1[j],kp2[des2arg]))

    return match[1]


```

```python
ke = matching(images[0], images[1])
```
